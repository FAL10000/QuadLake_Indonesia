```python
import polars as pl
from pathlib import Path
from urllib.request import urlretrieve
```

```python
source = pl.scan_csv('../source_csv/dataset-links.csv')
```

```python
indo_manifest = (
    source
    .filter(pl.col('Location').eq('Indonesia'))
    .select(["Location", "QuadKey", "Url", "Size", "UploadDate"])
    .collect()
)
```

```python
records = indo_manifest.to_dicts()

raw_dir = Path("../data/raw/")
raw_dir.mkdir(parents=True, exist_ok=True)

for row in records:
    quadkey = str(row["QuadKey"])
    url = row["Url"]
    upload_date = row["UploadDate"]

    filename = f"Indonesia_{quadkey}_{upload_date}.csv.gz"
    out_path = raw_dir / filename

    urlretrieve(url, out_path)
    print(f"Downloaded: {out_path}")
```
